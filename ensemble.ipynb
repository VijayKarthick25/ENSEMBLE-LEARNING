```python
import pandas as pd
test_set=pd.read_csv("/content/test_preprocessed.csv")
train_set=pd.read_csv("/content/train_preprocessed.csv")
test_set
```

```python
train_set
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_set.drop("Survived",axis=1),train_set["Survived"],test_size=0.2,random_state=42)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
def evaluation(model,x_train,y_train,x_test,y_test,train=True):
  if train:
    print("train result:\n")
    # print("accuracyscore:{0:4f}\n".format())
    print("classificationreport:\n",(classification_report(y_train,model.predict(x_train))))
    print("confusionmatrix:\n",(confusion_matrix(y_train,model.predict(x_train))))
  elif train==False:
     print("test result:\n")
     print("classificationreport:\n",classification_report(y_test,model.predict(x_test)))
     print("confusionmatrix:\n",confusion_matrix(y_test,model.predict(x_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtclassifier=DecisionTreeClassifier()
dtclassifier.fit(x_train,y_train)
evaluation(dtclassifier,x_train,y_train,x_test,y_test,train=True)
```

```python
evaluation(dtclassifier,x_train,y_train,x_test,y_test,train=False)
```

```python
from sklearn.ensemble import BaggingClassifier
bgclassifier=BaggingClassifier(estimator=dtclassifier,n_estimators=1000,bootstrap=True,n_jobs=1,random_state=42)
bgclassifier.fit(x_train,y_train)
```

```python
from sklearn.ensemble import StackingClassifier
estimators=[
    ("decisiontree",dtclassifier),
     ("randomclassifier",rfclassifier)
]

stac=StackingClassifier(
    estimators=estimators,
    final_estimator=lreg,
    passthrough=False
)
stac.fit(x_train,y_train)

```
